## Text to Image

```python
import openai
import base64
from PIL import Image
import io
from IPython.display import display

client = openai.OpenAI(
    api_key="sk-lbvVTILP1EQ0LLgZwxyGqQ",
    base_url="https://apidev.navigatelabsai.com"
)

response = client.images.generate(
    model="imagen-4.0-generate-001",
    prompt="A futuristic Learning environment where students are learning about sports",
    size="1024x1024"
)

image_base64 = response.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

image = Image.open(io.BytesIO(image_bytes))

display(image)
```

## Image to Text

```python
import openai
import base64

client = openai.OpenAI(
    api_key="sk-lbvVTILP1EQ0LLgZwxyGqQ",
    base_url="https://apidev.navigatelabsai.com"
)

image_url = "https://images.unsplash.com/photo-1565787154274-c8d076ad34e7?q=80&w=765&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D"

response = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "Describe this image"
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": image_url
                    }
                }
            ]
        }
    ]
)

print(response.choices[0].message.content)
```

## Text to Speech

```python
from openai import OpenAI
from IPython.display import Audio, display

client = OpenAI(
    api_key="sk-lbvVTILP1EQ0LLgZwxyGqQ",
    base_url="https://apidev.navigatelabsai.com"
)

response = client.audio.speech.create(
    model="gpt-4o-mini-tts",
    voice="alloy",
    input="Being Consistent Matters."
)

audio_bytes = response.read()

display(Audio(audio_bytes, autoplay=True))
```
